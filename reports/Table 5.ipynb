```python
import pandas as pd

df = pd.read_csv("df_clustered.csv")

df.head()
```

```python
df.columns
```

```python
features = [
    "Danceability", "Energy", "Loudness",
    "Speechiness", "Acousticness",
    "Instrumentalness", "Valence"
]

cluster_table = (
    df.groupby("Cluster_Name")[features]
      .mean()
      .round(3)
)

cluster_counts = df["Cluster_Name"].value_counts()

cluster_table["n"] = cluster_counts

cluster_table
```
