```python
import re
import pandas as pd
from langdetect import detect, LangDetectException
```

```python
df = pd.read_csv("Cleaned_lyrics_all.csv").dropna(subset=["lyrics"])
```

```python
def detect_lang(text):
    try:
        return detect(str(text)[:500])
    except LangDetectException:
        return "unknown"

df["language"] = df["lyrics"].apply(detect_lang)
df["language"].value_counts().head(10)
```

```python
df_en_es = df[df["language"].isin(["en", "es"])][["id", "Title", "Artists_All", "language", "lyrics"]]
df_en_es.shape
```

```python
df_en_es.to_csv("lyrics_en_es.csv", index=False)
```
